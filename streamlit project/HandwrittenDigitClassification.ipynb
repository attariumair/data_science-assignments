```python
from tensorflow.keras.datasets import mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Input

network = Sequential([
    Input(shape=(28, 28)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

network.compile(optimizer='adam',
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy'])

network.fit(train_images, train_labels, epochs=5)
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Create a blank 28x28 white canvas
img = Image.new("L", (28, 28), color=255)

plt.imshow(img, cmap='gray')
plt.title("Blank Image")
plt.axis("off")
plt.show()
```

```python
# Convert to numpy array
grayscale_img = img.convert('L')
grayscale_array = np.array(grayscale_img)

# Normalize (0–1)
grayscale_array = grayscale_array / 255.0

# Reshape to (1, 28, 28)
grayscale_array = grayscale_array.reshape(1, 28, 28)
```

```python
prediction = network.predict(grayscale_array)
predicted_digit = np.argmax(prediction)

plt.imshow(grayscale_img, cmap='gray')
plt.title(f"Predicted Digit: {predicted_digit}")
plt.axis("off")
plt.show()

print("Predicted Digit:", predicted_digit)
```

```python
network.save('digit_model.h5')
print("✅ Model saved successfully as 'digit_model.h5'")
```

```python
from tensorflow.keras.datasets import mnist
import numpy as np

(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Normalize (0–1)
train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Flatten, Dense

network = Sequential([
    Input(shape=(28, 28)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

network.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
network.fit(train_images, train_labels, epochs=5, validation_split=0.1)
```

```python
test_loss, test_acc = network.evaluate(test_images, test_labels)
print("Test Accuracy:", test_acc)
```

```python
network.save("digit_model.h5")
print("✅ Model saved successfully!")
```
