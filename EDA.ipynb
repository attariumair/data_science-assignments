# Step 4: Exploring Data Characteristics

In this step, we explore the characteristics of key numerical variables to understand their distribution, central tendency, spread, and detect any anomalies or outliers.

**Selected Columns for Analysis:**
- Sales
- Profit
- Quantity
- Discount

We will:
1. Compute summary statistics (mean, median, mode, std, etc.)
2. Assess skewness and kurtosis.
3. Visualize each column using box plots, histograms, and QQ plots.
4. Detect outliers using Z-score and IQR methods.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Select key numeric columns
num_cols = ['Sales', 'Profit', 'Quantity', 'Discount']
data = df[num_cols]

data.head()
```

```python
summary_stats = pd.DataFrame({
    'Mean': data.mean(),
    'Median': data.median(),
    'Mode': data.mode().iloc[0],
    'Std Deviation': data.std(),
    'Variance': data.var(),
    'Range': data.max() - data.min(),
    'IQR': data.quantile(0.75) - data.quantile(0.25),
    'Skewness': data.skew(),
    'Kurtosis': data.kurt()
}).round(2)

summary_stats
```

```python
plt.figure(figsize=(12,8))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data[col], color='skyblue')
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 2, i)
    sns.histplot(data[col], kde=True, color='orange')
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

```python
import scipy.stats as stats
import pylab

plt.figure(figsize=(12,8))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 2, i)
    stats.probplot(data[col], dist="norm", plot=pylab)
    plt.title(f'QQ Plot of {col}')
plt.tight_layout()
plt.show()
```

```python
# Calculate Z-scores
z_scores = np.abs(stats.zscore(data))
outliers_z = (z_scores > 3).sum()

outliers_z_df = pd.DataFrame({
    'Column': num_cols,
    'Outliers (Z>3)': outliers_z
})
outliers_z_df
```

```python
outliers_iqr = {}
for col in num_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers_iqr[col] = data[(data[col] < lower) | (data[col] > upper)][col].count()

outliers_iqr_df = pd.DataFrame(list(outliers_iqr.items()), columns=['Column', 'Outliers (IQR)'])
outliers_iqr_df
```

# Step 6: Handling Outliers

Outliers are data points that deviate significantly from other observations in the dataset.  
They can result from errors, unusual behavior, or genuine rare events.

In this step, we will:
1. Identify outliers using statistical methods (IQR and Z-score).
2. Decide whether to remove, adjust, or retain them.
3. Discuss their potential impact on analysis and modeling.


```python
num_cols = ['Sales', 'Profit', 'Quantity', 'Discount']

def detect_outliers_iqr(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(f"{col}: {len(outliers)} outliers detected ({(len(outliers)/len(df))*100:.2f}%)")
    return outliers

iqr_outliers = {}
for col in num_cols:
    iqr_outliers[col] = detect_outliers_iqr(df, col)
```

```python
from scipy import stats
import numpy as np

zscore_outliers = {}
for col in num_cols:
    z_scores = np.abs(stats.zscore(df[col]))
    outliers = df[z_scores > 3]
    print(f"{col}: {len(outliers)} outliers detected using Z-score")
    zscore_outliers[col] = outliers
```

```python
plt.figure(figsize=(12,8))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=df, y=col, color='skyblue')
    plt.title(f"Boxplot of {col}")
plt.tight_layout()
plt.show()
```

```python
def winsorize_series(s, lower_quantile=0.05, upper_quantile=0.95):
    lower = s.quantile(lower_quantile)
    upper = s.quantile(upper_quantile)
    return np.clip(s, lower, upper)

df_winsorized = df.copy()
for col in ['Sales', 'Profit']:
    df_winsorized[col] = winsorize_series(df[col])

print("Winsorization applied to Sales and Profit columns.")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12,8))

sns.boxplot(y=df['Sales'], ax=axes[0,0], color='salmon')
axes[0,0].set_title("Before Winsorization - Sales")

sns.boxplot(y=df_winsorized['Sales'], ax=axes[0,1], color='lightgreen')
axes[0,1].set_title("After Winsorization - Sales")

sns.boxplot(y=df['Profit'], ax=axes[1,0], color='salmon')
axes[1,0].set_title("Before Winsorization - Profit")

sns.boxplot(y=df_winsorized['Profit'], ax=axes[1,1], color='lightgreen')
axes[1,1].set_title("After Winsorization - Profit")

plt.tight_layout()
plt.show()
```
